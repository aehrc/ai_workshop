# Indexing and iteration

```python
for i in a_nice_list:
    print(i)
```

```python
for i in range(10):
    print(i)
```

```python
for keys in what_is_this:
    print(keys)
```

```python
for values in what_is_this.values():
    print(values)
```

```python
for keys, values in what_is_this.items():
    print(keys, values)
```

# Conditions

```python
score = 6
```

```python
if score > 5:
    print("You passed!")
```

```python
if score > 8:
    print("You passed!")
else:
    print("You failed!")
```

```python
if score > 6:
    print("You passed!")
elif score == 6:
    print("You barely passed!")
else:
    print("You failed!")
```

# List comprehension

```python
a = ['a', 'b', 'c', 'd', 'e']

accepted = [i for i in a]

accepted
```

```python
accepted = [i for i in a if i != 'c']
accepted
```

# Functions

```python
def a_function():
    print("Hello, World!")
```

```python
a_function()
```

```python
def a_function_with_parameters(name, age):
    print(f"Hello, {name}! You are {age} years old.")
```

```python
a_function_with_parameters("Hasan", 30)
```

```python
def one_that_returns_a_value(x, y):
    return x + y
```

```python
one_that_returns_a_value(5, 10)
```
